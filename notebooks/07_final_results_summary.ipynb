```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from IPython.display import Image, display

# Set visual style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Load Cleaned Data

```python
df = pd.read_csv('../data/cleaned/diwali_sales_cleaned.csv')
print(f"Dataset Shape: {df.shape}")
df.head()
```

## 2. Key Business Metrics
Summary of total revenue, orders, and customer base.

```python
metrics = pd.Series({
    "Total Revenue": f"‚Çπ{df['Amount'].sum():,.0f}",
    "Total Orders": f"{df['Orders'].sum():,}",
    "Unique Customers": f"{df['User_ID'].nunique():,}",
    "Avg Order Value": f"‚Çπ{df['Amount'].sum() / df['Orders'].sum():,.2f}"
})

print("--- High-Level KPIs ---")
print(metrics)
```

## 3. Visual Results (Saved Charts)
Below are the charts generated during the automated run.

```python
image_paths = [
    "../images/07_monthly_sales_trend.png",
    "../images/02_state_revenue.png",
    "../images/11_customer_funnel.png"
]

for path in image_paths:
    if os.path.exists(path):
        print(f"\nDisplaying: {os.path.basename(path)}")
        display(Image(filename=path))
    else:
        print(f"File not found: {path}")
```

## 4. Top Demographics (Quick Code Analysis)

```python
df.groupby(['Gender', 'Age Group'])['Amount'].sum().sort_values(ascending=False).unstack().plot(kind='bar')
plt.title('Revenue by Gender and Age Group')
plt.ylabel('Total Revenue (‚Çπ)')
plt.show()
```
